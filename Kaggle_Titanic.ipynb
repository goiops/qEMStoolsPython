# **4. Cleaning the dataset**


```python
 #complete missing age with median
for dataset in data_cleaner:
    dataset['Age'].fillna(dataset['Age'].median(), inplace = True)

# Complete missing null fare with median
for dataset in data_cleaner:
    dataset['Fare'].fillna(dataset['Fare'].median(), inplace = True)
    
# Complete missing null embarked with the mode
for dataset in data_cleaner:
    dataset['Embarked'].fillna(dataset['Embarked'].mode()[0], inplace = True) #mode needs an extra [0]

print('done')

#print a summary of the table columns that have a null values
print('Train columns with null values:\n', data1.isnull().sum())  
print('-'*10)


# print a summary of the table columns in the test dataset that have null values
print('\nTest columns with null values:\n', test_data.isnull().sum())
print('-'*10)

# So remaining column 'cabin' still has null data with 687 out of 891 with nulls. In that case we should just drop it. We drop also passenger Id and ticket reference

drop_columns = ['PassengerId','Cabin','Ticket']
data1.drop(drop_columns, axis = 1, inplace = True)
print(data1.columns)


# Now you can see that all columns have non-nulls. Therefore it is a clean training dataset
print('-'*10)
print(data1.isnull().sum())
```

# **5. Feature Engineering**

```python
print(data1.head(20))
```

```python
# Add additional columns for analysis

for dataset in data_cleaner:  #creating a new column of information on the family size
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1
    
    dataset['IsAlone'] = 1 # initialise column 'IsAlone to value of 1 as a booleon for True
    dataset['IsAlone'].loc[dataset['FamilySize'] > 1] = 0   #If boolean family size is greater than one, then it is false
    dataset['Title'] = dataset['Name'].str.split(",", expand = True)[1].str.split(".", expand = True)[0]
    
    ## Next create frequency bins for fare
    dataset['FareBin'] = pd.qcut(dataset['Fare'],4)  #qcut discretisation based on quantiles. 
    
    ## Next create frequency bins for age
    dataset['AgeBin'] = pd.cut(dataset['Age'].astype(int),5)  #cut discretises based on bins
    
    
```

```python
for dataset in data_cleaner:
    print(dataset.info())
    print(dataset.head(30))
    

print(data1['Title'].value_counts())  #summary stats of this categorical data column. As results per below, lots of additional tail titles. 
```

```python
# Cleaning up rare title names

stat_min = 10 #create a variable for minimum count for stat significance
title_names = data1['Title'].value_counts() < stat_min

print(title_names)
data1['Title'] = data1['Title'].apply(lambda x: 'Misc' if title_names.loc[x] == True else x)

print(data1['Title'].value_counts())

```

```python
# 1. Convert Formats where categorical data are converted to dummy variables for mathematical analysis using the Sklearn Label ENcoder. THis is to 
# convert categorical data or text data into numbers that the predictive modeller can better understand

# 2. In addition, this step will define the x (independant features) and the y (dependant/target) variables for data modelling

label = LabelEncoder()
for dataset in data_cleaner: 
    dataset['Sex_Code'] = label.fit_transform(dataset['Sex']) # this converts male and female into digits
    dataset['Embarked_Code'] = label.fit_transform(dataset['Embarked'])
    dataset['Title_Code'] = label.fit_transform(dataset['Title'])
    dataset['FareBin_Code'] = label.fit_transform(dataset['FareBin'])
    dataset['AgeBin_Code'] = label.fit_transform(dataset['AgeBin'])
    
# define the target variable y
target = ['Survived']

# Define the x variables for orginal features aka feature selection
data1_x = ['Sex','Pclass','Embarked','Title','SibSp','Parch','Age','Fare','FamilySize','IsAlone']

data1_x_calc = ['Sex_Code','Pclass','Embarked_Code','Title_Code','Parch','Age','Fare']

data1_xy = target + data1_x
print('Original X Y: ', data1_xy,'\n')

# Define  variables for original with bin features to remove continuous variables
data1_x_bin = ['Sex_Code', 'Pclass', 'Embarked_Code', 'Title_Code', 'FamilySize', 'AgeBin_Code', 'FareBin_Code']

data1_xy_bin = target + data1_x_bin
print('Bin X Y: ', data1_xy_bin, '\n')

# define x and y variable for dummy features
data1_dummy = pd.get_dummies(data1[data1_x])
data1_x_dummy = data1_dummy.columns.tolist()
data1_xy_dummy = target + data1_x_dummy
print('Dummy X Y: ', data1_xy_dummy, '\n')

print(data1_dummy.head())

```
